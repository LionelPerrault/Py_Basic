# f-string


### f-string is a formatting method introduced in python 3.6.<br>
### f-string is more efficient that formatting with .format() and format specifiers.

## __Syntax__: f"hello {}"

### Values are to be placed inside {} paranthesis.

```python
age = 20  # Declaring a integer value.

f_string = f"Hello I am {age} old"  # Declaring a formatted string with the help of f-string.

print(f_string)
```
